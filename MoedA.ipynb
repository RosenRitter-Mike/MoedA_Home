### Part Alef

1. Row customer data useage
   
   Pros:
   
      - The data is true to the source, with no edits. So no operation on the part of the analyst can be responsible for any oddities or             bias. (Examples: making the .mode() value for sex/gender as default for .fillna() may cause Linda to be Male or Johnatan to be       Female, Making the selery by gender statistic inacurate. Duplicates removed may be indeed part of the data and included on purpuse.)
     
   Cons:

    - The data may (most probably will) include mistakes, missing values, duplicates, etc. So the reliability may be problematic one spelling mistake, a dot forgotten or a zero in the wrong place and [John, age: 34, Salery: 14000] turns into [Joan, age:3.4, salery: 140000], ...) 

2. The mistakes that may appear are: missing values, duplicates, and spelling/typing mistakes.

   They can be delt with by removal(.drop_duplicates(), .dropna(), etc), filling them (.fillna(), etc) or if non essential data ignoring the data as non essential.

   We consider the case and make the decision.

3. Problem:
   
   The date is missing:
    One possible solution is if the customers are survayed based on address or customer_id, is to find other customers from the area or with ids close to the one with the missing date and fill based on that data. Othervise if the survay is Quarterly/Annual find what quarter or year is missing in this customers data. 
   
   The Address is missing:
    One possible solution is if the customers are survayed based on address, is to find other customers from the same survay date with the missing address and fill based on that data. Othervise if the survay is Quarterly/Annual find what was writen before in the customers data (previous surveys of that customer).
   
  The customer_id is duplicated:
    If its intentional ie surveys are periodical  Quarterly/Annual that makes sense same customer diferent time, or if address is different (a large custumer like Tnuva may have more then one site). Otherwise this data needs to be droped even if ststisticaly significant.
    
  The grade is missing (3 possible solutions or more):
  1. mean or median grade based on the area/address (it is logical to assume all the water to say Lod came from the same batch)
  2. mean or median grade based on the date of survey (we can assume that water from the same period of botteling came to the customers, so any problems are seen periodicaly).
  3. mean or median based on previous survays of this custumer (in the food industry periodical survays (qurterly/annual) of the same corprete customer are common, I work in the food industry I know...)
  4. mean or median grade based on general population
  5. just .mode() that grade (we primates are simple we tend to follow the herd, if everyone says the water is good we psicologicaly are programed to think it is even if it isn't).
